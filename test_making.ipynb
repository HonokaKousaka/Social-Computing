```python
import pandas as pd
import numpy as np

file_path = 'comment_20240626225604.jsonl'
df = pd.read_json(file_path, lines=True)

contents = df['content'].tolist()
```

```python
import re

for i in range(len(contents)):
    if contents[i].startswith("å›å¤@") or contents[i].startswith("å›å¤ @"):
        pattern_reply = r'å›å¤.*?:'
        contents[i] = re.sub(pattern_reply, '', contents[i])
    pattern_urls = r'http[s]?://\S+'
    contents[i] = re.sub(pattern_urls, '', contents[i])
    pattern = r'#.*?#'
    contents[i] = re.sub(pattern, '', contents[i])
```

```python
bias_sentence = []
for i in contents:
    if "å¥³æ‹³" in i and len(bias_sentence) < 100:
        bias_sentence.append(i)
```
